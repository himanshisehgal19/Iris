#  Prediction of Iris Species

```python
# import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the dataset

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
inputcols=['SepalLengthCm' ,'SepalWidthCm','PetalLengthCm','PetalWidthCm']
outputcols=['Species']
```

# Split the dataset

```python
split=int(0.7*df.shape[0])
train_data=df[:split]
test_data=df[split:]
test_data=test_data.reset_index(drop=True)
print(train_data.shape,test_data.shape)
```

```python
x=train_data[inputcols]
y=train_data[outputcols]
xtest=test_data[inputcols]
ytest=test_data[outputcols]
```

# Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier as ds
```

```python
sktree=ds(criterion='entropy',max_depth=5)
gktree=ds(criterion='gini',max_depth=5)
```

```python
sktree.fit(x,y)
```

```python
gktree.fit(x,y)
```

```python
sktree.score(xtest,ytest)
```

```python
gktree.score(xtest,ytest)
```

# Visualise Decision Tree with Entropy as a criterion

```python
from sklearn import tree
fig = plt.figure(figsize=(10,10))
tree.plot_tree(sktree,filled=True)
```
