# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the Dataset

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
inputcols=['SepalLengthCm' ,'SepalWidthCm','PetalLengthCm','PetalWidthCm']
outputcols=['Species']
```

```python
x=df[inputcols]
y=df[outputcols]
```

# Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier as ds
```

```python
sktree=ds(criterion='entropy')
gktree=ds(criterion='gini')
```

```python
sktree.fit(x,y)
gktree.fit(x,y)
```

# Visualize Decision Tree

```python
from sklearn import tree
fig = plt.figure(figsize=(25,20))
tree.plot_tree(sktree,filled=True)
```
